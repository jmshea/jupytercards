# Example Usage of Jupyter Cards

```python
from jupytercards import display_flashcards
github='https://raw.githubusercontent.com/jmshea/foundations-of-data-science-with-python/main/04-probability1/flashcards/'
```

## Basic usage, loading from URL

```python
display_flashcards(github + 'outcomes-samplespaces-events.json')
```

```python
# Demonstrate shuffling cards
display_flashcards(github + 'outcomes-samplespaces-events.json', shuffle_cards=True)
```

```python
# Test turning off known/not-known widgets
display_flashcards(github + 'outcomes-samplespaces-events.json', shuffle_cards=True, known_widgets=False)
```

```python
# Test local storage (persistence across sessions)
display_flashcards(github + 'outcomes-samplespaces-events.json', local_storage=True)
```

## Basic usage, load from file

```python
display_flashcards('example.json', local_storage=True)
```

# Basic Usage, Load from Python List of Dicts

Note that the data in this example introduces an optional third field for each card: the "topic". How to use this is shown further below.

```python
cards =[ 
    {'front': 'identity matrix',
     'back': r'A diagonal matrix in which the off-diagonal entries are all zero ' + 
             r'and the diagonal entries are all equal to 1. ' +
             r'The m×m identity matrix is denoted Im. ',
     'topic':'linear algebra'
    },
    {
        "front": "<i>average</i> (or <i>sample mean</i>) of a data set",
        "back": "The value that minimizes the squared error to the data. " +
                r"For a data set d0,d1,…,dN−1 the average is " +
                r"d―=1N∑i=0N−1di.",
        "topic" : "statistics"
    },
    {
        "front": "expected value<br>(discrete random variable)",
        "back": r"μX=E[X]=∑xPX(x).",
        "topic" : "probability"
    },
    {
        "front": "data",
        "back": "Collections of measurements, characteristics, or facts about a group."
    }
]
```

```python
display_flashcards(cards)
```

## More Advanced Usage, Selecting Subsets by Category

```python
display_flashcards(cards, topics='linear algebra')
```

```python
display_flashcards(cards, topics='statistics')
```

```python
display_flashcards(cards, topics=['probability', 'statistics'])
```

```python
display_flashcards(cards, topics=['probability', 'statistics'], local_storage=True)
```

## Test alternate color scheme

```python
display_flashcards(cards, topics=['probability', 'statistics'],
                   front_colors='jupytercon', back_colors='jupytercon')
```
